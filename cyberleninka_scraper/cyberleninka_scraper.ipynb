```python
from selenium import webdriver
import time, random
import requests
import bs4
import os, re
import time
```

```python
# here follows the directory with a text file for each domani
# in each of these text files there is a list of journals (which is, a list of journals for each domain)
# these journals are listed in a numeral bulletted list, below we proceed with the cleaning of these lists:
# we want to isolate every journal name

# after that, we create a dictionary "journal_dict"
# domain names will be keys
# for each key, a list of the journal names related to a domain.

journal_lst_dir = "./list_of_journals"
journal_dict = {file.strip('.txt') : [] for file in os.listdir(journal_lst_dir)}
```

```python
# here we create anothe dictionary:
# keys --> domain names
# values --> lists with link to articles

article_links_dict = {}
driver = webdriver.Firefox(executable_path='C:\Program Files\geckodriver\geckodriver.exe')

def article_link_collector(dict_journalLINKS=dict_journalLINKS):
    
    for journal_LINK_key in dict_journalLINKS:
        
        for journal_link in dict_journalLINKS[journal_LINK_key]:
        
            driver.get(journal_link)
            # here we want to know the number of years in we the journal published something
            # in the sites years are in the form of buttons
            years = driver.find_elements_by_css_selector("div.content > div > div.infoblock.visible > div > div.filter > ul button > span")
            
            # we want articles only from 2005 to 2019
            if len(years) > 15:
                del years[15:]
        
            for i in range(1, len(years)):
                # on the journal page there is a button that, when clicked, the buttons for years is shown
                # so we click on it
                year_selector = driver.find_element_by_css_selector("div.content > div > div.infoblock.visible > div > div.filter > div > button")
                year_selector.click()
            
                # here we select the "year" element that, when clicked, will show the links for articles of that year
                year_elem = driver.find_element_by_css_selector("div.content > div > div.infoblock.visible > div > div.filter > ul > li:nth-child({}) > button > span".format(i))
                year_elem.click()
                
                # there is another step:
                # in journal pages articles are divided into series, each year has different series
                tom_elems = driver.find_elements_by_css_selector("ul > li.active > button > span")
                
                # if we have only a series per year we won't have to bother clicking on each series' button
                # before getting the article links
                if len(tom_elems) == 1:
                    article_elems = driver.find_elements_by_css_selector('a[href^="/article/n"]')
                    article_elemsLINK = [article_elem.get_attribute("href") for article_elem in article_elems if article_elem]
                
                    for link in article_elemsLINK:
                        if not journal_LINK_key in article_links_dict:
                            article_links_dict[journal_LINK_key] = []
                            article_links_dict[journal_LINK_key].append(link)
                        else:
                            article_links_dict[journal_LINK_key].append(link)
                
                # when there's more than one series, instead, we will have to click on the series button before getting
                # the various article links (all the series button are ordered by number, here's the use of "j")
                # after clicking on the series button, we proceed getting the article links
                else:
                    for j in range(1, len(tom_elems)):
                        tom_selector = driver.find_element_by_css_selector("#div.content > div > div.infoblock.visible > div > div.filter > ul > li:nth-child({}) > button > span".format(i))
                        tom_selector.click()
                        
                        article_elems = driver.find_elements_by_css_selector('a[href^="/article/n"]')
                        article_elemsLINK = [article_elem.get_attribute("href") for article_elem in article_elems if article_elem]
                
                        for link in article_elemsLINK:
                            if not journal_LINK_key in article_links_dict:
                                article_links_dict[journal_LINK_key] = []
                                article_links_dict[journal_LINK_key].append(link)
                            else:
                                article_links_dict[journal_LINK_key].append(link)
    
    driver.close()
    return article_links_dict
                    
article_link_collector()
```
